# AutoRegresive Integrated Moving Average

```python
plt.figure(figsize=(10,5))
plt.plot(air)
```

```python
best_order = None 
best_aic = np.inf
```

```python
for p in range(7, 15):
    for d in range(5):
        for q in range(7,15):
            try:
                model=ARIMA(air, order=(p, d, q))
                result = model.fit()
                if result.aic < best_aic:
                    best_aic = result.aic
                    best_order = (p, d, q)
            except:
                continue
                
```
